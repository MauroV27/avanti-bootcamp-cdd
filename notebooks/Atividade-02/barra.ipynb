# Gráfico de barras - Dataset de arrecadação por filmes : 

### Import das bibliotecas

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

### Carregamento dos dados de arrecadação dos filmes : 

```python
df_movies = pd.read_csv(
"https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv" )
df_movies.head() # ver como os dados estão colocados no dataset
```

### Gráfico de barras sem nenhum tratamento na visualização dos dados ( forma errada de apresentar informação ) :

```python
ax_movies = sns.barplot(data=df_movies, y="arrecadacao", x="titulo")
```

### Gráfico após tratamento dos dados e estilização : 

```python
# Cria nova coluna com os valores em milhão
df_movies = df_movies.assign(arrecadacao_milhao = df_movies.arrecadacao / 1_000_000 ) 

# Sort dos dados do dataset com base na arrecadação. Por padrão ele ordena do menor para o maior, por isso foi usado o ascending=False
df_movies = df_movies.sort_values(by="arrecadacao", ascending=False)

# Configuração do gráfico : 
sns.set_style("whitegrid") # linhas verticais

plt.figure(figsize=[5, 2])

ax_movies = sns.barplot(data=df_movies, x="arrecadacao_milhao", y="titulo", color="tomato")

ax_movies.bar_label(
    ax_movies.containers[0],
    fmt="%.1f",
    label_type="center",
    color="white"
)

plt.xlabel("Arrecadação ( em milhões de dolares )")
plt.ylabel("Nome do filme")
plt.title("Arrecadação por filmes")
plt.show()
```
