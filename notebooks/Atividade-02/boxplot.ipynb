# Gráfico de histograma - Dataset de tips : 

### Import das bibliotecas

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

### Carregamento dos dados :

```python
df_tips = sns.load_dataset("tips")

df_tips.head()
```

### Gráfico sem nenhum tratamento na visualização dos dados ( forma errada de apresentar informação ) :

```python
ax_tips = sns.boxplot(data=df_tips, x="day", y="total_bill")
```

### Gráfico após tratamento dos dados e estilização : 

```python
sns.set_style("whitegrid")

plt.figure(figsize=[6, 3])

# gráfico orientado pelo material
ax_tips = sns.boxplot(data=df_tips, x="day", y="total_bill", color="cornflowerblue")

plt.xlabel("Dia da semana")
plt.ylabel("Conta (USD)")
plt.title("Distribução de gorjetas ao longo da semana")

plt.show()
```

#### Gráfico da relação entre gorjetas com total de conta por dia da semana :

Para um melhor entendimento da relação entre a gorjeta paga pelos clientes foi feita uma analise com base em uma nova variavel 'percentage_tip_by_total', que descreve a relação entre uma gorjeta e o total pago pelo cliente ( incluindo a gorjeta ).

```python
df_tips = df_tips.assign( percentage_tip_by_total = df_tips.tip / ( df_tips.total_bill + df_tips.tip) )
df_tips.head()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=[6, 3])

# gráfico criado com base na nova variavel : percentage_tip_by_total
ax_tips = sns.boxplot(data=df_tips, x="day", y="percentage_tip_by_total", color="cornflowerblue", showmeans=True, 
                      meanprops={"marker":"+",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"white",
                      "markersize":"5"})

plt.xlabel("Dia da semana")
plt.ylabel("Gorjeta sob valor total pago pelos clientes")
plt.title("Relação entre gorjetas/total pago pelo cliente ao longo da semana")

plt.show()
```
