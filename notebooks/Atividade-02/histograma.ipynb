# Gráfico de histograma - Dataset de penguins : 

### Import das bibliotecas

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

### Carregamento dos dados :

```python
df_penguins = sns.load_dataset("penguins")

df_penguins.head()
```

### Gráfico sem nenhum tratamento na visualização dos dados ( forma errada de apresentar informação ) :

```python
ax_penguins = sns.histplot(data=df_penguins, x="body_mass_g")
```

### Gráfico após tratamento dos dados e estilização : 

```python
sns.set_style("whitegrid")

ax_penguins = sns.histplot(data=df_penguins, x="body_mass_g")

ax_penguins.set(frame_on=False)
ax_penguins.axhline(0, color="k", clip_on=False)

plt.ylabel("Quantidade")
plt.xlabel("Massa (g)")
plt.title("Distribuição de massa dos pinguins")

plt.show()
```
