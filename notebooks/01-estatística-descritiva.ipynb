```python
vars_quali_map = [ ["Variavel", "Valor"]]

for var in titanic_qualitative_vars["variavel"].to_list():
    unique_values = df_titanic[var].unique()
    for v in unique_values:
        vars_quali_map.append( [var, v])
        
display(pd.DataFrame(vars_quali_map))
```
